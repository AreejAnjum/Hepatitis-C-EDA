## Explodatory Data Analysis (EDA)

## Gender 

```python

#`Plot a countplot with value labels`
ax=sns.countplot(x='Sex', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
  


```

```python
#Plot a barplot for Sex vs Alt level 
total_victoms= df.groupby(["Sex"], as_index=False)['ALT'].count().sort_values(by='Sex', ascending=False)
sns.barplot(x='Sex', y='ALT', data=total_victoms)
```

Hence from above it is clear tha male has high level of alt than woman

```python
ax=sns.countplot(x='Category', data=df, hue='Sex')
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
for bars in ax.containers:
    ax.bar_label(bars)
    
#to avoid title overlapping with x axis labels
#plt.figure(figsize=(10,5))
#sns.countplot(data=df, y="Category", hue="Sex")
#plt.tight_layout()
```

```python
#to convert a continuous variable into categorical variable using quartiles
#df["AgeGroup"] = pd.qcut(df["Age"], q=4, labels=["Q1","Q2","Q3","Q4"])
#df.head()
```

```python
bins = [0, 19, 29, 39, 49, 59, 69, 79, 120]
labels = ["0-19","20-29","30-39","40-49","50-59","60-69","70-79","80+"]

df["AgeGroup"] = pd.cut(df["Age"], bins=bins, labels=labels, right=True, include_lowest=True)

df.head(100)
```

```python
# Drop the AgeGroup_Q column
df = df.drop(columns=['AgeGroup_Q'])
df
```

```python
chol_Count = df.groupby(['AgeGroup'], as_index=False)['CHOL'].count().sort_values(by='CHOL', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='AgeGroup', y='CHOL', data=chol_Count )
```

From above graph, its clear that people age between 40-49 are having CHOL level

```python
ax=sns.countplot(x='AgeGroup', data=df)
sns.set(rc={'figure.figsize':(10,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

From above analysis, we can conclude that most people having Hepatitis age between 40-49
